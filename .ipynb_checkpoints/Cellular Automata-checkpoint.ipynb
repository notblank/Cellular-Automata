```python
import numpy as np 
import pandas as pd
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
def opinion(u0, T):
    N = len(u0)
    
    u = np.zeros((T, N))
    
    u[0, :] = u0
    
    for j in np.arange(1, T):
        
        if((u0[1] + u0[0] + u0[N-1]) > 1):
            u[j,0] = 1
        else:
            u[j,0] = 0

        for i in np.arange(1, N-1):
            if((u0[i+1] + u0[i] + u0[i-1]) > 1):
                u[j, i] = 1
            else:
                u[j, i] = 0
        
        if((u0[1] + u0[N-1] + u0[N-2]) > 1):
            u[j, N-1] = 1
        else:
            u[j, N-1] = 0

            
        u0 = u[j, :]

    return(u)
```

```python
u0 = np.zeros(14)
index = np.arange(0, 14, 2)
u0[index] = 1
u0[5] = 1
u0
```

```python
O = opinion(u0, 10)
plt.matshow(O[0:4,:] , cmap=plt.cm.viridis)
plt.xticks([])
plt.yticks([])
plt.savefig('three friends.png', bbox_inches='tight')
plt.show()
```

```python
def traffic(u0, T):
    N = len(u0)
    
    u = np.zeros((T, N))
    
    u[0, :] = u0
    
    for j in np.arange(1, T):
        
        for i in np.arange(1, N-1):
            if(u0[i-1] == 0 and u0[i] == 0 and u0[i+1] == 0):
                u[j, i] = 0
            elif(u0[i-1] == 1 and u0[i] == 0 and u0[i+1] == 0):
                u[j, i] = 1
            elif(u0[i-1] == 0 and u0[i] == 1 and u0[i+1] == 0):
                u[j, i] = 0
            elif(u0[i-1] == 0 and u0[i] == 0 and u0[i+1] == 1):
                u[j, i] = 0
            elif(u0[i-1] == 1 and u0[i] == 1 and u0[i+1] == 0):
                u[j, i] = 0
            elif(u0[i-1] == 1 and u0[i] == 0 and u0[i+1] == 1):
                u[j, i] = 1
            elif(u0[i-1] == 0 and u0[i] == 1 and u0[i+1] == 1):
                u[j, i] = 1
            elif(u0[i-1] == 1 and u0[i] == 1 and u0[i+1] == 1):
                u[j, i] = 1
            
            if(u0[N-1] == 0 and u0[0] == 0 and u0[1] == 0):
                u[j, 0] = 0
            elif(u0[N-1] == 1 and u0[0] == 0 and u0[1] == 0):
                u[j, 0] = 1
            elif(u0[N-1] == 0 and u0[0] == 1 and u0[1] == 0):
                u[j, 0] = 0
            elif(u0[N-1] == 0 and u0[0] == 0 and u0[1] == 1):
                u[j, 0] = 0
            elif(u0[N-1] == 1 and u0[0] == 1 and u0[1] == 0):
                u[j, 0] = 0
            elif(u0[N-1] == 1 and u0[0] == 0 and u0[1] == 1):
                u[j, 0] = 1
            elif(u0[N-1] == 0 and u0[0] == 1 and u0[1] == 1):
                u[j, 0] = 1
            elif(u0[N-1] == 1 and u0[0] == 1 and u0[1] == 1):
                u[j, 0] = 1
                
            if(u0[N-2] == 0 and u0[N-1] == 0 and u0[0] == 0):
                u[j, N-1] = 0
            elif(u0[N-2] == 1 and u0[N-1] == 0 and u0[0] == 0):
                u[j, N-1] = 1
            elif(u0[N-2] == 0 and u0[N-1] == 1 and u0[0] == 0):
                u[j, N-1] = 0
            elif(u0[N-2] == 0 and u0[N-1] == 0 and u0[0] == 1):
                u[j, N-1] = 0
            elif(u0[N-2] == 1 and u0[N-1] == 1 and u0[0] == 0):
                u[j, N-1] = 0
            elif(u0[N-2] == 1 and u0[N-1] == 0 and u0[0] == 1):
                u[j, N-1] = 1
            elif(u0[N-2] == 0 and u0[N-1] == 1 and u0[0] == 1):
                u[j, N-1] = 1
            elif(u0[N-2] == 1 and u0[N-1] == 1 and u0[0] == 1):
                u[j, N-1] = 1

        u0 = u[j, :]

    return(u)
```

```python
T = traffic(u0, 11)
plt.matshow(T, cmap=plt.cm.viridis)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
def fredkin(u0, T):
    N = len(u0)
    
    u = np.zeros((T, N))
    
    u[0, :] = u0
    
    for j in np.arange(1, T):
        
        if(u0[1] == u0[N-1]):
            u[j,0] = 0
        else:
            u[j,0] = 1

        for i in np.arange(1, N-1):
            if(u0[i+1] == u0[i-1]):
                u[j, i] = 0
            else:
                u[j, i] = 1
        
        if(u0[1]== u0[N-2]):
            u[j, N-1] = 0
        else:
            u[j, N-1] = 1

            
        u0 = u[j, :]

    return(u)
```

```python
u_center = np.zeros(20)
u_center[7:11] = 1
u_center
```

```python
F = fredkin(u_center , 5)
plt.matshow(F , cmap=plt.cm.viridis)
plt.xticks([])
plt.yticks([])
plt.savefig('A copy.png', bbox_inches='tight')
plt.show()
```
